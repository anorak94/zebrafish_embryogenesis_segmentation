```python
def plot_histogram_of_extrema_min_max(r):
    """
    since we have a small finite signal instead of characterizing it by 
    psd, autocorrelation etc. i collect statistics on the values of 
    r_max and r_min
    
    """
    r_max = find_vals_of_r_extrema(r, "max")
    r_min = find_vals_of_r_extrema(r, "min")
    
    f, (ax1, ax2) = plt.subplots(1, 2)

    ax1.hist(r_max, bins=15, density=True, alpha=0.6, color='r')
    ax2.hist(r_min, bins=15, density=True, alpha=0.6, color='g')

def listRightIndex(alist, value):
    return len(alist) - alist[-1::-1].index(value) -1

def closest_time(u, u_0, t):
    vals = list((u-u_0)**2)
    val = min(vals)
    return int(listRightIndex(vals, val))
    
def find_vals_of_r_extrema(r2d, extrema_type=None):
    """
    computes r_max and r_min
    """
    if (extrema_type == "max"):
        extrema  = signal.argrelextrema(r2d, np.greater)
    else:
        extrema  = signal.argrelextrema(r2d, np.less)
    return r[extrema]



def dist_between_peaks(peaks, dx):
    """
    computes distance between peaks the wavelength gives the length of the 
    somites, each somite is roughly 5 to 6 cells so a check on the veracity 
    of the simulations would be this result
    """
    dist = []
    for i in range(1, len(peaks)):
        dist.append(2*dx*(peaks[i]-peaks[i-1]))
    dist = np.asarray(dist)
    stats = np.unique(dist, return_index=False, return_inverse=False, return_counts=True, axis=None)
    return stats, peaks, dist


def extract_somite_statistics_from_readout(r2d, dx):
    """
    extracts relevant statistics from the readout pattern interleaves the 
    min and max vals into one single peak list
    
    """
    peaks, _ = signal.find_peaks(r2d)
    extrema_g,  = signal.argrelextrema(r2d, np.greater)
    extrema_l,  = signal.argrelextrema(r2d, np.less)
    np.append(extrema_l, 0)
    if extrema_g[0]<extrema_l[0]:
        peaks =np.asarray( list(itertools.chain(*zip(list(extrema_g), list(extrema_l))))[:-1])
    else:
        peaks =np.asarray( list(itertools.chain(*zip(list(extrema_l), list(extrema_g))))[:-1])

    return dist_between_peaks(peaks, dx)

def return_heaviside(n_c, n_t, dx, dt, v_0):
    """
    computes the heaviside function which simulates the wavefront 
    could have vectorized it but the run time is not too large 
    as number of cells and time steps of simulation is very small 
    """
    ha2d = np.ones((n_c, n_t))
    for i in range(n_t):
        for j in range(n_c):
            ha2d[j][i] = int((j*dx - v_0*i*dt)>0)
    return ha2d

def plot_fft_of_readout(r, x):
    """
    plots the fourier transform of the readout to see how regular the pattern is 
    use it even though the signal is finite 
    """
    fft_r = fft(r)
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(x,fft_r)
    ax.set_title("fft of r vs x")





def plot_peaks_on_readout(r2d, peaks, x_, dx):
    """
    plots the peaks on the readout to see if the correct statistics 
    have been collected, plots additional ticks so i can just read out 
    the wavelength instead of running the statistics collection functions
    """
    f, ax = plt.subplots()
    ax.plot(x_, r2d)
    ax.scatter(peaks*dx, r2d[peaks], c = "r")
    plt.xticks(np.arange(min(x_), max(x_)+1, 1.0))

    
def plot_individual_cell(w, u, r, h, t, x, loc):
    """
    given the various variables plots them for a specified single cell
    so that i look at the various parameters values without running the simulation 
    again and again
    """
    
    f, (ax) = plt.subplots(2, 2, figsize = (18 ,8))
        
    ax[0, 0].plot(t, h[loc, :], c = "k")
    ax[0, 0].set_title("heaviside vs t")
    ax[0, 1].plot(t, u[loc, :], c = "r")
    ax[0, 1].set_title("u vs t")
    ax[1, 0].plot(t, w[loc, :], c = "g")
    ax[1, 0].set_title("w vs t")
    ax[1, 1].plot(x, r, c = "b")
    ax[1, 1].set_title("r vs x")
    
def get_oscillations_of_w_single_cell(w2d, t):
    """
    returns statistics on w , the max and min values of w
    and the threshold times to go from w_min to w_max and w_max to w_min
    """
    peaks, _ = signal.find_peaks(w2d)
    extrema_g,  = signal.argrelextrema(w2d, np.greater)
    extrema_l,  = signal.argrelextrema(w2d, np.less)
    w_max = w2d[extrema_g]
    w_min = w2d[extrema_l]
    t_p_max = t[extrema_g]
    t_max = np.asarray([x - t_p_max[i - 1] for i, x in enumerate(t_p_max)][1:])
    t_p_min = t[extrema_l]
    t_min = np.asarray([x - t_p_min[i - 1] for i, x in enumerate(t_p_min)][1:])
    np.append(extrema_l, 0)
    if extrema_g[0]<extrema_l[0]:
        peaks =np.asarray( list(itertools.chain(*zip(list(extrema_g), list(extrema_l))))[:-1])
    else:
        peaks =np.asarray( list(itertools.chain(*zip(list(extrema_l), list(extrema_g))))[:-1])

    return w_max, w_min, t_max, t_min


def run_sim_wout_coupling_and_noise(dic, plot=False, plot_r = False, plot_mollifier=False):
    """
    integrates and runs the equations has flags for plotting 
    again could have written vectorized equations but speed is not 
    too much of a concern with the small steps plus i add coupling (c) later 
    which i could have still vectorized by using np.roll and shifting the w
    matrix forward and backward as i do with the mollifier function maybe later fix 
    """
    u2d = np.ones((dic["n_c"], dic["c"]*dic["n_t"]))
    w2d = np.zeros((dic["n_c"], dic["c"]*dic["n_t"]))
    r2d = np.zeros(dic["n_c"])
    t = np.asarray([i*dic["dt"] for i in range(dic["n_t"]*dic["c"])])
    x_ = np.asarray([i*dic["dx"] for i in range(dic["n_c"])])
    ha2d = return_heaviside(dic["n_c"], dic["n_t"]*dic["c"], dic["dx"], dic["dt"], dic["v_0"])
    
    for i in range(dic["c"]*dic["n_t"]-1):
        for cell in range(1, dic["n_c"]-1):
            
            u2d[cell][i+1] = u2d[cell][i] + dic["dt"]*(-u2d[cell][i]*(1/dic["tau_2"])+ (1/dic["tau_2"])*ha2d[cell][i])
            
            if (i - dic["tau_d"] >= 0):
                c = dic["beta"]*(w2d[cell-1][i-dic["tau_c"]]+w2d[cell+1][i-dic["tau_c"]])
                w2d[cell][i+1] = w2d[cell][i] + dic["dt"]*(-w2d[cell][i]*(1/dic["tau_w"]) + dic["gamma1"]*(u2d[cell][i]**dic["n"]/(u2d[cell][i]**dic["n"]+dic["K_u"]**dic["n"]))*(1/(1+w2d[cell][i-dic["tau_d"]]**dic["m"])) + c)
            else:
                c = dic["beta"]*(w2d[cell-1][i-dic["tau_c"]]+w2d[cell+1][i-dic["tau_c"]])
                w2d[cell][i+1] = w2d[cell][i] + dic["dt"]*(-w2d[cell][i]*(1/dic["tau_w"]) + dic["gamma1"]*(u2d[cell][i]**dic["n"]/(u2d[cell][i]**dic["n"]+dic["K_u"]**dic["n"]))*(1)+c)
    
    
    
    for cell in range(1, dic["n_c"]-1):
        """
        mollifier : since i have a dirac delta in integration i handle that by defining a 
        very peaked distribution around the value of u_0 which is a parameter in my simulation 
        
        """
        mollifier = define_distribution_around_u0(u2d[cell], param_dic["u_0"])
        
        if (plot_mollifier):
            if (cell == param_dic["loc"]):
                plt.plot(u2d[cell], mollifier)
                plt.title("mollifier vs u")
        for time in range(1, dic["c"]*dic["n_t"]-1):
            increment = w2d[cell][time]*mollifier[time]
            r2d[cell]  = r2d[cell]+increment
    
    if (plot):
        f, (ax) = plt.subplots(2, 2, figsize = (18 ,8))
        
        ax[0, 0].plot(t, ha2d[dic["loc"], :], c = "k")
        ax[0, 0].set_title("heaviside vs t")
        ax[0, 1].plot(t, u2d[dic["loc"], :], c = "r")
        ax[0, 1].set_title("u vs t")
        ax[1, 0].plot(t, w2d[dic["loc"], :], c = "g")
        ax[1, 0].set_title("w vs t")
        ax[1, 1].plot(x_, r2d, c = "b")
        ax[1, 1].set_title("r vs x")
    if (plot_r):
        plt.figure(figsize=(6, 4))
        plt.plot(x_, r2d)
        plt.title("r vs x")
    return u2d, w2d, r2d, ha2d, t, x_


def define_distribution_around_u0 (u, u_0, t=param_dic["n_t"]*param_dic["c"]):
    """
    smooth dirac delta approximation by defining a peaked gaussian around the value 
    of u_0
    """
    uis = np.linspace(0, 1, t)
    val = norm.pdf(uis, u_0, 1e-4)
    val_max = np.amax(val)
    mollifier = val/val_max
    ind_closest_point = closest_time(u, u_0, t)
    ind_1 = np.where(mollifier == 1)[0][0]
    shift = ind_closest_point - ind_1
    moll_2 = np.roll(mollifier, shift)
    return moll_2
```

```python
param_dic["v_0"] = .25
param_dic["dt"] = 0.5
param_dic["n_t"]  = 200
param_dic["n_c"] = 50

param_dic["delay"] = 11
param_dic["tau_d"] = round(param_dic["delay"]/param_dic["dt"])
param_dic["u_0"] =.75
param_dic["c"] = 15
param_dic["tau_w"] = 1.
param_dic["gamma1"] = 21
param_dic["dx"] = 1.0
param_dic["tau_c"] = 0
param_dic["tau_2"] = 20
param_dic["beta"] = 0.05
param_dic["n"] = 4
param_dic["m"] = 5
param_dic["K_u"] = 0.3

param_dic["loc"] = 15


u, w, r, h, t, x = run_sim_wout_coupling_and_noise(param_dic, plot = True, plot_r = False, plot_mollifier=False)
```
